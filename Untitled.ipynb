```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler().fit(df)
scaled=scale.transform(df)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

n_clusters =[4,5,6,7,8,9]

for K in n_clusters:
    cluster=KMeans(n_clusters=K ,random_state=10)
    predict =cluster.fit_predict(scaled)
    
    score=silhouette_score(scaled,predict,random_state=10)
    print("For n_clusters = {}, silhouette_score is {}".format(K,score))
```

```python
# building a K-Means model for K = 8
model = KMeans(n_clusters= 8, random_state= 10)
model.fit(scaled)
```
